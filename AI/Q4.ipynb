```
while True:
    print("\nMenu:")
    print("1. Add two numbers")
    print("2. Subtract two numbers")
    print("3. Exit")

    choice = input("Enter your choice: ")
    if choice == '1':
        a = float(input("Enter first number: "))
        b = float(input("Enter second number: "))
        print("Result:", a + b)

    elif choice == '2':
        a = float(input("Enter first number: "))
        b = float(input("Enter second number: "))
        print("Result:", a - b)

    elif choice == '3':
        print("Exiting")
        break

    else:
        print("Invalid")
```
